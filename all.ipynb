```python
def astar(start,stop):
  open=set(start)
  close=set()
  g={}
  parent={}
  g[start]=0
  parent[start]=start
  while(len(open)>0):
    n=None
    for v in open:
      if(n==None or g[v]+h(v)<g[n]+h(n)):
        n=v
    if(n==stop or graph[n]==None):
      pass
    else:
      for (m,w) in getneigh(n):
        if m not in open and m not in close:
          open.add(m)
          g[m]=g[n]+w
          parent[m]=n
        else:
          if(g[m]>g[n]+w):
            g[m]=g[n]+w
            parent[m]=n
            if m in close:
              close.remove(m)
              open.add(m)
    if n==None:
        print("path does not exists")
        return None
    if n==stop:
        path=[]
        while(parent[n]!=n):
          path.append(n)
          n=parent[n]
        path.append(start)
        path.reverse()
        print(path)
        return path
    open.remove(n)
    close.add(n)
  print("path does not exists")
  return None
```

```python
def pr():
  for x in board:
    for y in x:
      print(y,end="|")
    print()

def win(p):
  for x in board:
    if x.count(p)==3:
      return True
  for x in range(3):
    k=0
    for y in range(3):
      if board[y][x]==p:
        k+=1
    if k==3:
      return True
  k,j=0,0
  for x in range(3):
    if board[x][x]==p:
      k+=1
    if board[x][2-x]==p:
      j+=1
  if k==3 or j==3:
    return True
  return False

def draw(): 
  for x in board:
    for y in x:
      if y=='-':
        return False
  return True

def computer():
  for i in range(3):
    for j in range(3):
      if board[i][j]=='-':
        board[i][j]='O'
        if win('O'):
          return
        board[i][j]='-'
  for i in range(3):
    for j in range(3):
      if board[i][j]=='-':
        board[i][j]='X'
        if win('X'):
          board[i][j]='O'
          return
        board[i][j]='-'
  for i in range(3):
    for j in range(3):
      if board[i][j]=='-':
        board[i][j]='O'
        return
  
```

```python
board=[['-','-','-'],
       ['-','-','-'],
       ['-','-','-']]

while True:
  if draw():
    print("Match drawn")
    break
  if not win('X') and not win('O'):
    print("Player X turn")
    x,y=map(int,input().split())
    board[x-1][y-1]='X'
    pr()
  else:
    break
  if not win('X') and not win('O'):
    print("Computer turn")
    computer()
    pr()
  else:
    break
  
if win('X'):
  print("Player X won")
elif win('O'):
  print("Computer won")
```

```python
import numpy as np
def step(v):
  if(v>=0):
    return 1
  return 0

def perceptron(x,w,b):
  v=np.dot(w,x)+b
  return step(v)

def notl(x):
  w=-1
  b=0.5
  return perceptron(x,w,b)

def orl(x):
  w=np.array([1,1])
  b=-0.5
  return perceptron(x,w,b)

def andl(x):
  w=np.array([1,1])
  b=-1.5
  return perceptron(x,w,b)

def xorl(x):
  y1=andl(x)
  y2=orl(x)
  y3=notl(y1)
  fx=andl(np.array([y2,y3]))
  return fx

test1=np.array([0,0])
test2=np.array([0,1])
test3=np.array([1,0])
test4=np.array([1,1])

print("{}|{}={}".format(0,0,orl(test1)))
print("{}|{}={}".format(0,1,orl(test2)))
print("{}|{}={}".format(1,0,orl(test3)))
print("{}|{}={}".format(1,1,orl(test4)))

print("{}&{}={}".format(0,0,andl(test1)))
print("{}&{}={}".format(0,1,andl(test2)))
print("{}&{}={}".format(1,0,andl(test3)))
print("{}&{}={}".format(1,1,andl(test4)))

print("{}^{}={}".format(0,0,xorl(test1)))
print("{}^{}={}".format(0,1,xorl(test2)))
print("{}^{}={}".format(1,0,xorl(test3)))
print("{}^{}={}".format(1,1,xorl(test4)))

print("~{}={}".format(0,notl(0)))
print("~{}={}".format(1,notl(1)))
```

```python
n=int(input())
clear=[True]*(n+1)
on=[0]*(n+1)
goalon=[0]*(n+1)

def puton(x,y):
  if not clear[x]:
    putontable(on.index(x))
  if not clear[y]:
    putontable(on.index(y))
  clear[y]=False
  clear[on[x]]=True
  on[x]=y
  print(x," on ",y)

def putontable(x):
  if not clear[x]:
    putontable(on.index(x))
  clear[on[x]]=True
  on[x]=0
  print("put ",x," on table")

for i in range(1,n+1):
  print("Block-",i)
  pos=int(input())
  if pos==1:
    on[i]=0
  elif pos==2:
    on[i]=int(input())
    clear[i]=False

for i in range(1,n+1):
  print(i,clear[i],on[i])

for i in range(1,n+1):
  print("Block-",i)
  pos=int(input())
  if pos==1:
    goalon[i]=0
  elif pos==2:
    goalon[i]=int(input())

base=[]
if goalon!=on:
  for i in range(1,n+1):
    if(goalon[i]==0):
      if on[i]!=0:
        putontable(i)
      base.append(i)
  while (goalon!=on):
    b=base.pop(0)
    try:
      x=goalon.index(b)
      if on[x]!=b:
        puton(x,b)
      base.append(x)
    except:
      pass
```
